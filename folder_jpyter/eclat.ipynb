```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import random
```

## Data Preprocessing

```python
dataset = pd.read_csv('Market_Basket_Optimisation.csv', header = None)
transactions = []
for i in range(0, 7501):
  transactions.append([str(dataset.values[i,j]).strip() for j in range(0, 20)])
```

## Training the Eclat model on the dataset

```python
cust_id = []
normal = 5
for i in range(1,6501):
    val = str(i) 
    length = len(val)
    idx = normal-length
    cust_id.append('CUS_'+'0'*idx+val)
for i in range(1001):
    val = str(random.randint(1,6500))
    length = len(val)
    idx = normal-length
    cust_id.append('CUS_'+'0'*idx+val)
```

```python
dataset.insert(0,'Customer ID',cust_id)
```

```python
dataset.to_csv('purchase.csv',index=False)
```

```python
from apyori import apriori
rules = apriori(transactions = transactions, min_support = 0.0000000003, min_confidence = 0.2, min_lift = 3, min_length = 2, max_length = 2)
```

## Visualising the results

### Displaying the first results coming directly from the output of the apriori function

```python
results = list(rules)
```

```python
results
```

### Putting the results well organised into a Pandas DataFrame

```python
def inspect(results):
    lhs         = [tuple(result[2][0][0])[0] for result in results]
    rhs         = [tuple(result[2][0][1])[0] for result in results]
    supports    = [result[1] for result in results]
    return list(zip(lhs, rhs, supports))
resultsinDataFrame = pd.DataFrame(inspect(results), columns = ['Product 1', 'Product 2', 'Support'])
```

### Displaying the results sorted by descending supports

```python
import pickle
pickle.dump(resultsinDataFrame,open("results.pickle","wb"))
```
